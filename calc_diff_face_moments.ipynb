```python
def random_choice_5paths(vectors, batch_size=5):
    vectors_list = list(vectors)
    while True:
        random_vectors = random.sample(vectors_list, batch_size)
        yield random_vectors
```

```python
def distance(a, b):
    return np.sqrt(np.sum(np.square(np.array(a) - np.array(b))))
```

```python
def random_choose_5paths(vectors, batch_size=5):
    vector_list = list(vectors)
    while True:
        random_vectors = random.sample(vector_list, batch_size)
        yield random_vectors
```

```python
# 设置计算非本人脸的平均moments
# random_vectors: generated by @method random_choose_5paths
# random_vectors include 5 people, several faces each
def calc_other_moments(self_faces, random_vectors):
    self_face = random.sample(self_faces, 1)
    print(type(self_face))
    moments = []
    for random_vector in random_vectors:
        t_vector = random.sample(random_vectors, 1)
        print(type(t_vector))
        moments.append(distance(self_face, t_vector))
    return np.mean(moments)
```

```python
ovr2_faces_info = ovr2_faces_info[: 40]
ovr2_faces_info['random moments'] = ovr2_faces_info['descriptors'].apply(calc_other_moments, args=([next(random_choose_5paths(ovr2_faces_info['descriptors']))]))
```
